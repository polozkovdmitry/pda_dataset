## 1. Загрузка данных

```python
import pandas as pd

url = "https://github.com/polozkovdmitry/ada_hse_econ/raw/refs/heads/main/2026_spring/sem_1_intro/sessions_1_excel.csv"

df = pd.read_csv(url)
# df в конце ячейки - удобно для глаз
df
```

```python
print(df)

# а функция print - удобно для отладки (llm-friendly😊)
# нажми «копировать выходные данные ячейки»
```

```python
pd.set_option('display.expand_frame_repr', False)
print(df)

# теперь точно удобно
```

```python
print(df.head(5))
print('-'*10)
print(df.tail(5))
print('-'*10)
print(df.sample(5))
print('-'*10)
print(df.describe())
print('-'*10)
print(df.info())
```

```python
df['session_start_dttm'] = pd.to_datetime(df['session_start_dttm'])
# теперь столбец можно складывать и преобразовывать (e.g. взять только месяц)

# id не складываются - лучше от греха подальше сделать строкой
# чтобы самим потом не напороться
df['user_id'] = df['user_id'].astype(str)
```

```python
# _rand столбцы были техническими для генерации - удалим их

# на всякий случай оставим полную версию
df_full = df.copy()

# на подумать: почему нельзя код ниже в одной ячейке с этим?
```

```python
df = df_full.drop(columns=['_rand moscow', '_rand desktop', '_rand satisfaction'])
# всегда проверяем себя
print(df.columns)
```

## 3. Операции с столбцами

```python
# на следующем семинаре будем делить пользователей на новичков и бывалых
# на каждого юзера в общем случае несколько сессий
# надо найти время самой первой сессии

df['registered_dttm'] = (
    df.groupby('user_id')['session_start_dttm']
      .transform('min')
)
```

```python
# проверим глазами, что столбец создался правильно

print(df.sort_values(by=['user_id', 'session_start_dttm'])[['user_id', 'session_start_dttm', 'registered_dttm']].head(20))
```
