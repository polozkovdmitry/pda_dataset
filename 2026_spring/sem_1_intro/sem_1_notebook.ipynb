## 1. Загрузка данных

```python
import pandas as pd

url = "https://github.com/polozkovdmitry/ada_hse_econ/raw/refs/heads/main/2026_spring/sem_1_intro/sessions_3_full.xlsx"

df = pd.read_excel(url)
# df в конце ячейки - удобно для глаз
df
```

```python
df['session_start_dttm'] = pd.to_datetime(df['session_start_dttm'])
# теперь столбец можно складывать и преобразовывать (e.g. взять только месяц)

# id не складываются - лучше от греха подальше сделать строкой
# чтобы самим потом не напороться
df['user_id'] = df['user_id'].astype(str)
```

```python
# _rand столбцы были техническими для генерации - удалим их

# на всякий случай оставим полную версию
df_full = df.copy()

# на подумать: почему нельзя код ниже в одной ячейке с этим?
```

```python
df = df_full.drop(columns=['_rand device', '_rand group', '_rand metric'])
# всегда проверяем себя
print(df.columns)
```

## 3. Операции с столбцами

```python
# На следующем семинаре будем делить пользователей на новичков и бывалых
# На каждого юзера в общем случае несколько сессий
# Надо найти время самой первой сессии

df_registered_dttm = (
    df.groupby('user_id', as_index=True)['session_start_dttm']
      .min()
      .to_frame(name='registered_dttm')
).reset_index()

df = df.merge(
    df_registered_dttm,
    left_on='user_id',
    right_on='user_id',
    how='left'
)

# input('проверить assert на каждую таблицу что dates not out of range')
```
